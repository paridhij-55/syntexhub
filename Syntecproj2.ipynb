```python
#Student Performance Analysis Using Pandas
import pandas as pd

# Step 1: Read CSV file
df = pd.read_csv("/content/sample_data/pandas/students.csv")

# Step 2: Display first 5 rows
print("First 5 rows of data:")
print(df.head())

# Step 3: Display last 5 rows
print("\nLast 5 rows of data:")
print(df.tail())

# Step 4: Check data types
print("\nData Types:")
print(df.dtypes)

# Step 5: Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Mean, Median, Min, Max
print("\nMean Marks:", df["Marks"].mean())
print("Median Marks:", df["Marks"].median())
print("Minimum Marks:", df["Marks"].min())
print("Maximum Marks:", df["Marks"].max())
print("Total Students:", df["Marks"].count())

# Step 6: Select columns
print("\nSelected Columns (Name & Marks):")
print(df[["Name","Marks"]])

# Step 7: Filter students with marks > 80
high_marks = df[df["Marks"] > 80]

print("\nStudents scoring more than 80:")
print(high_marks)

# Step 8: Slice subset
print("\nFirst 3 rows:")
print(df.iloc[0:3])

# Step 9: Save filtered data
high_marks.to_csv("top_students.csv", index=False)

print("\nFiltered data saved as top_students.csv")
```
